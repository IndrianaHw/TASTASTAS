```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import geopandas as gpd
import contextily as ctx
```

```python
indo = gpd.read_file("map data/INDONESIA_PROP.shp")
indo.head()
```

```python
data = gpd.read_file("Stasiun1-point.shp")
data.head()
```

```python
# Reproject data to Web Mercator (required for contextily basemap)
data = data.to_crs(epsg=3857)

# Plot
fig, ax = plt.subplots(figsize=(10, 10))
data.plot(ax=ax, marker='o', color='blue', markersize=5, label="Stasiun")

# Add basemap
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)

# Set plot title and labels
plt.title("Plot Data Lokasi Stasiun di Atas Peta")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend()
plt.show()
```
