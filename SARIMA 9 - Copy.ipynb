```python
def test_stationarity(timeseries):
    #Determing rolling statistics
    MA = timeseries.rolling(window=28).mean()
    MSTD = timeseries.rolling(window=28).std()

    #Plot rolling statistics:
    plt.figure(figsize=(15,5))
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(MA, color='red', label='Rolling Mean')
    std = plt.plot(MSTD, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)

    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
def tsplot(y, lags=None, figsize=(12, 7), style='bmh'):
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
        
    with plt.style.context(style):    
        fig = plt.figure(figsize=figsize)
        layout = (2, 2)
        ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
        acf_ax = plt.subplot2grid(layout, (1, 0))
        pacf_ax = plt.subplot2grid(layout, (1, 1))
        
        y.plot(ax=ts_ax)
        p_value = sm.tsa.stattools.adfuller(y)[1]
        ts_ax.set_title('Time Series Analysis Plots\n Dickey-Fuller: p={0:.5f}'.format(p_value))
        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax)
        plt.tight_layout()
```

```python
import itertools
import numpy as np
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Misal: data sudah dalam format time series dan bernama data
# data harus berupa Series dengan index berupa datetime
# Contoh:
# data = pd.read_csv("yourfile.csv", index_col='Date', parse_dates=True)

# Parameter ranges
p = d = q = range(0, 3)       # (p,d,q) from 0 to 2
P = D = Q = range(0, 2)       # (P,D,Q) from 0 to 1
m = 28                        # Musim QBO

# Generate semua kombinasi parameter
pdq = list(itertools.product(p, d, q))
seasonal_pdq = list(itertools.product(P, D, Q, [m]))

# Untuk menyimpan hasil AIC
results = []

# Grid search
for param in pdq:
    for seasonal_param in seasonal_pdq:
        try:
            model = SARIMAX(data,
                            order=param,
                            seasonal_order=seasonal_param,
                            enforce_stationarity=False,
                            enforce_invertibility=False)
            results_fit = model.fit(disp=False)
            results.append({
                'order': param,
                'seasonal_order': seasonal_param,
                'AIC': results_fit.aic
            })
        except:
            continue

# Konversi ke DataFrame dan sortir
results_df = pd.DataFrame(results)
results_df = results_df.sort_values('AIC').reset_index(drop=True)

# Tampilkan model terbaik
print("Model SARIMA terbaik berdasarkan AIC:")
print(results_df.head(1))

# Jika ingin semua model disortir berdasarkan AIC
print("\nTop 5 model dengan AIC terendah:")
print(results_df.head(5))
```

```python
# Fit model SARIMA terbaik
model = SARIMAX(data,
                order=(2, 0, 2),
                seasonal_order=(0, 1, 1, 28),
                enforce_stationarity=False,
                enforce_invertibility=False)

model_fit = model.fit()

# Ringkasan model
print(model_fit.summary())
```

```python
# --- Split data: 70% train, 20% test, 10% forecast ---
n = len(data)
n_train = int(n * 0.7)
n_test = int(n * 0.2)
n_forecast = n - n_train - n_test  # Sisa data

train = data.iloc[:n_train]
test = data.iloc[n_train:n_train+n_test]
forecast = data.iloc[n_train+n_test:]

# --- Model fitting di data train (contoh: ARIMA/SARIMA) ---
# Ganti model_fit sesuai model yang kamu pakai
# Misal: from statsmodels.tsa.statespace.sarimax import SARIMAX
# model = SARIMAX(train, order=(1,0,1), seasonal_order=(1,0,1,12))
# model_fit = model.fit(disp=False)

# Untuk contoh, saya pakai dummy model yang hanya mengulang nilai terakhir train
class DummyModel:
    def predict(self, start, end, dynamic=False):
        return np.repeat(train.values[-1], end - start + 1)
model_fit = DummyModel()

# --- Prediksi training dan testing ---
train_pred = model_fit.predict(start=0, end=len(train)-1, dynamic=False)
test_pred = model_fit.predict(start=len(train), end=len(train)+len(test)-1, dynamic=False)
forecast_pred = model_fit.predict(start=len(train)+len(test), end=n-1, dynamic=False)

# --- Plot training dan testing ---
fig, axs = plt.subplots(3, 1, figsize=(10, 12))  

# Plot Train
axs[0].plot(train.index, train.values, label='Data Training QBO', color='blue')
axs[0].plot(train.index, train_pred, label="Prediksi Training", color='red', linestyle='--')
axs[0].set_title('Plot Training')
axs[0].legend()
axs[0].grid(True)

# Plot Test
axs[1].plot(test.index, test.values, label='Data Testing QBO', color='green')
axs[1].plot(test.index, test_pred, label="Prediksi Testing", color='red', linestyle='--')
axs[1].set_title('Plot Testing')
axs[1].legend()
axs[1].grid(True)

# Plot Forecast
axs[2].plot(forecast.index, forecast.values, label='Data Aktual Forecast', color='black')
axs[2].plot(forecast.index, forecast_pred, label='Prediksi Forecast', color='orange', linestyle='--')
axs[2].set_title('Plot Forecast 10%')
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()

# --- Evaluasi ---
# Pastikan dalam array 1D
y_test_actual = np.array(test).flatten()
y_test_pred = np.array(test_pred).flatten()

y_forecast_actual = np.array(forecast).flatten()
y_forecast_pred = np.array(forecast_pred).flatten()

# R2, MAE, MSE, RMSE untuk test dan forecast
mse_test = mean_squared_error(y_test_actual, y_test_pred)
rmse_test = np.sqrt(mse_test)
mae_test = mean_absolute_error(y_test_actual, y_test_pred)
r2_test = r2_score(y_test_actual, y_test_pred)

mse_forecast = mean_squared_error(y_forecast_actual, y_forecast_pred)
rmse_forecast = np.sqrt(mse_forecast)
mae_forecast = mean_absolute_error(y_forecast_actual, y_forecast_pred)
r2_forecast = r2_score(y_forecast_actual, y_forecast_pred)

metrics_qbo = pd.DataFrame({
    'Metric': ['R²', 'MAE', 'MSE', 'RMSE'],
    'Test Data': [r2_test, mae_test, mse_test, rmse_test],
    'Forecast Data': [r2_forecast, mae_forecast, mse_forecast, rmse_forecast]
}).round(4)

print(metrics_qbo)

# --- Scatter plot test ---
plt.figure(figsize=(6, 5))
plt.scatter(y_test_actual, y_test_pred, alpha=0.7)
plt.plot([y_test_actual.min(), y_test_actual.max()], [y_test_actual.min(), y_test_actual.max()], color='black', linestyle='--')
plt.xlabel("Actual values (Test)")
plt.ylabel("Predicted values (Test)")
plt.title("Scatter Plot of Actual vs Predicted (Testing Data)")
plt.text(0.05, 0.90, f"R² = {r2_test:.2f}", transform=plt.gca().transAxes)
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Scatter plot forecast ---
plt.figure(figsize=(6, 5))
plt.scatter(y_forecast_actual, y_forecast_pred, alpha=0.7)
plt.plot([y_forecast_actual.min(), y_forecast_actual.max()], [y_forecast_actual.min(), y_forecast_actual.max()], color='black', linestyle='--')
plt.xlabel("Actual values (Forecast)")
plt.ylabel("Predicted values (Forecast)")
plt.title("Scatter Plot of Actual vs Predicted (Forecast Data)")
plt.text(0.05, 0.90, f"R² = {r2_forecast:.2f}", transform=plt.gca().transAxes)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- ANALISIS SPEKTRAL QBO vs RESIDUAL SARIMA ---

# Anggap `data_qbo` adalah array QBO aslinya
# Anggap `residuals` adalah residual dari SARIMA (sudah ada di kode-mu sebelumnya)
# Jika data_qbo belum ada, load sesuai kebutuhan (misal dari file atau variabel)

# Contoh (ganti sesuai datamu):
# data_qbo = np.loadtxt('data_qbo.txt')
# Jika data_qbo sudah berupa pandas Series/DataFrame, bisa langsung: data_qbo = df['QBO'].values

data_qbo = data['25 hPa'].values

sampling_interval = 1  # misal bulanan

def plot_periodogram(data, label, sampling_interval=1):
    fs = 1 / sampling_interval
    freqs, power = periodogram(data, fs=fs, scaling='spectrum')
    plt.plot(freqs, power, label=label)
    return freqs, power

plt.figure(figsize=(12,6))
freqs_qbo, power_qbo = plot_periodogram(data_qbo, 'QBO Asli')
freqs_res, power_res = plot_periodogram(residuals, 'Residual SARIMA')
plt.xlabel('Frekuensi (1/bulan)')
plt.ylabel('Power')
plt.title('Periodogram QBO Asli vs Residual SARIMA')
plt.legend()
plt.xlim(0, 0.5)
plt.grid()
plt.show()

# Cari periode dominan
def dominant_period(freqs, power):
    idx = np.argmax(power[1:]) + 1  # skip frekuensi nol
    dom_freq = freqs[idx]
    dom_period = 1 / dom_freq if dom_freq != 0 else np.inf
    return dom_period, dom_freq

dom_period_qbo, dom_freq_qbo = dominant_period(freqs_qbo, power_qbo)
dom_period_res, dom_freq_res = dominant_period(freqs_res, power_res)

print(f"Periode dominan QBO asli: {dom_period_qbo:.2f} bulan (frekuensi: {dom_freq_qbo:.4f})")
print(f"Periode dominan residual SARIMA: {dom_period_res:.2f} bulan (frekuensi: {dom_freq_res:.4f})")

# Interpretasi:
if np.abs(dom_period_qbo - dom_period_res) < 2:  # toleransi 2 bulan, sesuaikan jika perlu
    print("⚠️ Periode dominan residual SARIMA masih mirip dengan data asli. Model SARIMA belum sepenuhnya menghilangkan siklus utama QBO.")
else:
    print("✅ Periode dominan residual SARIMA berbeda dengan data asli. Model SARIMA sudah menangkap siklus utama QBO.")
```
